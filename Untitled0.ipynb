```python
from ultralytics import YOLO
import os
import cv2
import numpy as np

# Load YOLOv8x model
model = YOLO('yolov8x.pt')

# Run ByteTrack; retain results generator
results = model.track(
    source='clip.mp4',
    tracker='bytetrack.yaml',
    conf=0.3,
    iou=0.7,
    stream=True,    # streaming yields per-frame Results
    persist=True
)

# Create output folders
frames_dir = 'YOLOX_outputs/byte_frames'
tracks_dir = 'YOLOX_outputs/byte_exp'
os.makedirs(frames_dir, exist_ok=True)
os.makedirs(tracks_dir, exist_ok=True)

# Initialize list to accumulate track entries
all_tracks = []

# Iterate over streamed Results
for idx, res in enumerate(results, start=1):
    # res.im is the annotated image (numpy array)
    img = np.asarray(res.im)
    # Save frame as JPG
    frame_path = os.path.join(frames_dir, f'{idx:06d}.jpg')
    cv2.imwrite(frame_path, img)

    # Extract MOT-format lines for this frame
    # Tracks are in res.tracks with fields: id, xyxy, conf
    for track in res.tracks:
        tid = int(track.id.cpu().numpy())
        x1, y1, x2, y2 = track.xyxy.cpu().numpy().tolist()
        score = float(track.conf.cpu().numpy())
        # Format: frame, id, x, y, w, h, score, -1, -1, -1
        w, h = x2 - x1, y2 - y1
        line = f"{idx},{tid},{x1:.2f},{y1:.2f},{w:.2f},{h:.2f},{score:.2f},-1,-1,-1"
        all_tracks.append(line)

print(f"Saved {idx} annotated frames to {frames_dir}/")

# Save all tracks to a single MOT-format text file
tracks_path = os.path.join(tracks_dir, 'tracks.txt')
with open(tracks_path, 'w') as f:
    f.write("\n".join(all_tracks))

print(f"Saved MOT-format tracks to {tracks_path}")
print("Frames directory listing:", os.listdir(frames_dir)[:5], "...")  # sample listing
```

```python
from ultralytics import YOLO
import os
import cv2

# Load YOLOv8x model
model = YOLO('yolov8x.pt')

# Run ByteTrack; results is a generator of Results
results = model.track(
    source='clip.mp4',
    tracker='bytetrack.yaml',
    conf=0.3,
    iou=0.7,
    stream=True,
    persist=True
)

# Create output folders
frames_dir = 'YOLOX_outputs/byte_frames'
tracks_dir = 'YOLOX_outputs/byte_exp'
os.makedirs(frames_dir, exist_ok=True)
os.makedirs(tracks_dir, exist_ok=True)

all_tracks = []

# Iterate over streamed Results
for idx, res in enumerate(results, start=1):
    # 1) Annotate frame: plot() draws boxes and returns the image
    annotated = res.plot()  # numpy array (H,W,3) in RGB

    # Convert RGB to BGR for OpenCV and save frame
    bgr = cv2.cvtColor(annotated, cv2.COLOR_RGB2BGR)
    frame_path = os.path.join(frames_dir, f'{idx:06d}.jpg')
    cv2.imwrite(frame_path, bgr)

    # 2) Extract tracks for this frame
    for t in res.tracks:  # each track has id, xyxy, conf
        tid = int(t.id.cpu().numpy())
        x1, y1, x2, y2 = t.xyxy.cpu().numpy().tolist()
        score = float(t.conf.cpu().numpy())
        w, h = x2 - x1, y2 - y1
        line = f"{idx},{tid},{x1:.2f},{y1:.2f},{w:.2f},{h:.2f},{score:.2f},-1,-1,-1"
        all_tracks.append(line)

print(f"Processed and saved {idx} annotated frames in '{frames_dir}'")

# Save MOT-format tracks
tracks_path = os.path.join(tracks_dir, 'tracks.txt')
with open(tracks_path, 'w') as f:
    f.write("\n".join(all_tracks))

print(f"Saved MOT-format tracks to '{tracks_path}'")
print("Sample frames directory listing:", os.listdir(frames_dir)[:5])
```

```python
from ultralytics import YOLO
import os
import cv2

# 1. Clean up any existing frames
frames_dir = 'processed_frames'
if os.path.isdir(frames_dir):
    for f in os.listdir(frames_dir):
        if f.endswith('.jpg'):
            os.remove(os.path.join(frames_dir, f))
else:
    os.makedirs(frames_dir)

# 2. Load model and run ByteTrack in streaming mode
model = YOLO('yolov8x.pt')
results = model.track(
    source='clip.mp4',
    tracker='bytetrack.yaml',
    conf=0.3,
    iou=0.7,
    stream=True,
    persist=False
)

# 3. Process and save first 5 annotated frames and collect tracks
tracks = []
for idx, res in enumerate(results, start=1):
    # Save annotated frame
    img = res.plot()
    bgr = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    frame_path = os.path.join(frames_dir, f'{idx:06d}.jpg')
    cv2.imwrite(frame_path, bgr)
    print(f"Saved frame {idx} to '{frame_path}'")

    # Extract tracking info for this frame
    # Each res.boxes contains detection; for tracking we look at res.boxes.id
    # Convert to CPU numpy
    boxes = res.boxes.cpu().numpy()
    ids   = res.boxes.id.cpu().numpy()
    xyxy  = res.boxes.xyxy.cpu().numpy()
    confs = res.boxes.conf.cpu().numpy()

    for tid, (x1, y1, x2, y2), score in zip(ids, xyxy, confs):
        w, h = x2 - x1, y2 - y1
        tracks.append(f"{idx},{int(tid)},{x1:.2f},{y1:.2f},{w:.2f},{h:.2f},{score:.2f},-1,-1,-1")

    if idx >= 5:
        break

# 4. Save collected tracks to a text file
tracks_file = 'first5_tracks.txt'
with open(tracks_file, 'w') as f:
    f.write('\n'.join(tracks))
print(f"Saved tracking info for first 5 frames to '{tracks_file}'")

# 5. Verify
print("Frames directory listing:", sorted(os.listdir(frames_dir)))
with open(tracks_file) as f:
    print("Tracks text:\n", f.read())
```

```python
from ultralytics import YOLO
import os
import cv2

# 1. Prepare output directories
frames_dir = 'processed_frames'
if os.path.isdir(frames_dir):
    for f in os.listdir(frames_dir):
        if f.endswith('.jpg'):
            os.remove(os.path.join(frames_dir, f))
else:
    os.makedirs(frames_dir)

# 2. Load YOLOv8x model and run ByteTrack in streaming mode
model = YOLO('yolov8x.pt')
results = model.track(
    source='clip.mp4',
    tracker='bytetrack.yaml',
    conf=0.3,
    iou=0.7,
    stream=True,
    persist=False
)

# 3. Iterate over all frames, save annotated image, and collect tracking lines
tracks = []
for idx, res in enumerate(results, start=1):
    # Annotate frame
    img = res.plot()
    bgr = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    frame_path = os.path.join(frames_dir, f'{idx:06d}.jpg')
    cv2.imwrite(frame_path, bgr)

    # Extract track info from res.boxes
    boxes = res.boxes.cpu().numpy()
    ids   = res.boxes.id.cpu().numpy()
    xyxy  = res.boxes.xyxy.cpu().numpy()
    confs = res.boxes.conf.cpu().numpy()

    for tid, (x1, y1, x2, y2), score in zip(ids, xyxy, confs):
        w, h = x2 - x1, y2 - y1
        tracks.append(f"{idx},{int(tid)},{x1:.2f},{y1:.2f},{w:.2f},{h:.2f},{score:.2f},-1,-1,-1")

# 4. Save all tracks to MOT-format text file
tracks_file = 'all_frames_tracks.txt'
with open(tracks_file, 'w') as f:
    f.write('\n'.join(tracks))

# 5. Summary
print(f"Processed {idx} frames.")
print(f"Annotated frames saved to '{frames_dir}/'.")
print(f"Tracking data saved to '{tracks_file}'.")
```
